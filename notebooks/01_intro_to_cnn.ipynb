# 01 — Intro to CNNs

This notebook introduces image tensors, convolutions, feature maps, and a baseline CNN.

**Learning goals**
- Inspect CIFAR-10 batches
- Build intuition for tensor shapes
- Define and run a small CNN
- Connect architecture choices to image classification


```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(ROOT / "src") not in sys.path:
    sys.path.append(str(ROOT / "src"))

import torch
import matplotlib.pyplot as plt
from data_loader import DataConfig, get_cifar10_loaders
from model_cnn import SimpleCNN
```

## 1. Load a batch of images

CIFAR-10 contains 10 classes of 32x32 RGB images.


```python
train_loader, test_loader, classes = get_cifar10_loaders(DataConfig(batch_size=8, image_size=32))
images, labels = next(iter(train_loader))
print("images shape:", images.shape)
print("labels shape:", labels.shape)
print("classes:", classes)
```

```python
fig, axes = plt.subplots(2, 4, figsize=(10, 5))
for ax, img, label in zip(axes.flat, images, labels):
    img = img.permute(1, 2, 0) * 0.5 + 0.5
    ax.imshow(img)
    ax.set_title(classes[label])
    ax.axis("off")
plt.tight_layout()
```

## 2. Define the model

Read the CNN and map each block to a role:
- convolution = local pattern detector
- ReLU = nonlinearity
- pooling = downsampling
- classifier = decision layer


```python
model = SimpleCNN(num_classes=len(classes))
print(model)
```

```python
with torch.no_grad():
    logits = model(images)
print("logits shape:", logits.shape)
```

## 3. Shape tracing exercise

Change the input batch size or image size and observe how the output changes.

**Try this:**
1. Print the shape after each convolution block.
2. Replace max pooling with average pooling.
3. Increase channels and compare parameter count.


```python
num_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Total params: {num_params:,}")
print(f"Trainable params: {trainable_params:,}")
```
