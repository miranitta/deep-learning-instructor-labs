# 03 — Transfer Learning

Use a pretrained ResNet18 and compare it with a custom CNN.


```python
from pathlib import Path
import sys
ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(ROOT / "src") not in sys.path:
    sys.path.append(str(ROOT / "src"))

from model_resnet import build_resnet18
from model_cnn import SimpleCNN
```

```python
cnn = SimpleCNN(num_classes=10)
resnet = build_resnet18(num_classes=10, pretrained=True, freeze_backbone=True)
```

```python
def count_params(model):
    total = sum(p.numel() for p in model.parameters())
    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    return total, trainable

for name, model in [("cnn", cnn), ("resnet", resnet)]:
    total, trainable = count_params(model)
    print(name, f"total={total:,}", f"trainable={trainable:,}")
```

## Fine-tuning discussion

Common strategies:
1. Freeze all backbone layers and train only the classifier.
2. Unfreeze the last residual block.
3. Fine-tune the full network with a smaller learning rate.


```python
# Example: unfreeze the last residual block
for param in resnet.layer4.parameters():
    param.requires_grad = True

count_params(resnet)
```
