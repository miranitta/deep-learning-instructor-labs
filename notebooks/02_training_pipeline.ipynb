# 02 — Training Pipeline

This notebook walks through optimization, loss tracking, and validation.


```python
from pathlib import Path
import sys
ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(ROOT / "src") not in sys.path:
    sys.path.append(str(ROOT / "src"))

import torch
from torch import nn, optim
from data_loader import DataConfig, get_cifar10_loaders
from model_cnn import SimpleCNN
from train import train_one_epoch, eval_one_epoch
from utils import get_device, set_seed
```

```python
set_seed(42)
device = get_device()
train_loader, test_loader, classes = get_cifar10_loaders(DataConfig(batch_size=64))
model = SimpleCNN(num_classes=len(classes)).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)
device
```

## Run a short training job

Use a small number of epochs to inspect the pipeline quickly.


```python
history = []
for epoch in range(1, 3):
    train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, device)
    val_loss, val_acc = eval_one_epoch(model, test_loader, criterion, device)
    history.append((epoch, train_loss, train_acc, val_loss, val_acc))
    print(epoch, train_loss, train_acc, val_loss, val_acc)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(history, columns=["epoch", "train_loss", "train_acc", "val_loss", "val_acc"])
df
```

```python
plt.figure(figsize=(6,4))
plt.plot(df["epoch"], df["train_loss"], label="train_loss")
plt.plot(df["epoch"], df["val_loss"], label="val_loss")
plt.legend()
plt.xlabel("epoch")
plt.ylabel("loss")
plt.title("Training curve")
plt.show()
```
